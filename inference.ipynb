# Inference pipeline for testing the model
The following example illustrates how to perform damage classification on an image test subset which belongs to the same wind turbines as the training data, and on a test subset of images belonging to new wind turbines. Using:
- The Inception Resnet V1 Neural Network with a Siamese architecture
- The Inception Resnet V1 Neural Network pretained with the VGGface2 dataset

```python
from facenet_pytorch import InceptionResnetV1, fixed_image_standardization
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import numpy as np
import pandas as pd
import os
import shutil
import matplotlib.pyplot as plt 
import seaborn as sns
from tensorflow.keras.preprocessing import image
from PIL import Image
from sklearn.manifold import TSNE
from functions import Siamese

workers = 0 if os.name == 'nt' else 4
```

### Baseline Experiment: Inception Resnet V1 pretained with VGGface2

```python
resnet = torch.load('C:/Users/U339700/Documents/Palas/models/baseline/baseline.pth').eval()
resnet
```

Compute all images embeddings and visualize them

```python
def collate_fn(x):
    return x[0]

trans = transforms.Compose([
    transforms.Resize((256, 256)),
    np.float32,
    transforms.ToTensor(),
    fixed_image_standardization
])

data_dir = "C:/Users/U339700/Documents/Palas/dataset/siamesas_resize_train"
dataset = datasets.ImageFolder(data_dir, transform=trans)
dataset.idx_to_class = {i:c for c, i in dataset.class_to_idx.items()}
loader = DataLoader(dataset, collate_fn=collate_fn, num_workers=workers)
```

```python
dataset.idx_to_class
```

```python
aligned = []
classes = []

for x, y in loader:
     aligned.append(x)
     classes.append(dataset.idx_to_class[y])
     
aligned = torch.stack(aligned).to(device)
```

```python
batch_size=100
batches = len(aligned)//batch_size
len_read = (batches-1)*batch_size+batch_size

embs = []

if len_read < len(aligned): # Si el dataset es mayor que el batch size
    for batch in range(batches):
        print('Batch ', batch+1, ' / ', batches+1)
        embeddings = resnet(aligned[batch*batch_size:batch*batch_size+batch_size]).detach().cpu()
        embs.append(embeddings.tolist())
        # tsne.append(TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs)))
        # embs = []
        
    # Imgenes ltimo batch
    print('Batch ', batches+1, ' / ', batches+1)
    embeddings = resnet(aligned[len_read:len(aligned)]).detach().cpu()
    embs.append(embeddings.tolist())
    # tsne.append(TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs)))
else:
    embeddings = resnet(aligned).detach().cpu()
    # tsne.append(TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs)))

tsne_emb = []
for b, emb in enumerate(embs):
    for v in emb:
        tsne_emb.append(v)
len(tsne_emb)
```

```python
tsne = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(tsne_emb))
tsne_df = pd.DataFrame()

tsne_df['x'] = tsne[:, 0]
tsne_df['y'] = tsne[:, 1]
tsne_df['class'] = classes
tsne_df
```

```python
sns.set_theme()
sns.lmplot('x', 'y', data=tsne_df, hue='class', fit_reg=False, legend=False, palette="Set2", scatter_kws={"s": 20, 'alpha':0.7})
plt.legend(title='Class', loc='best', labels=['D-0', 'D-2', 'D-3', 'D-4 and D-5'], facecolor='White')
plt.title('Baseline model image embeddings', fontsize = 14)
plt.show()
```

### Siamese Resnet Experiment: Siamese architecture with Inception Resnet V1 trained with our training data

Compute all images embeddings and visualize them

```python
net = Siamese(
    classify=False,
    pretrained='vggface2',
    embs=True
    ).eval()
net.load_state_dict(torch.load('C:/Users/U339700/Documents/Palas/models/proposed/best-model-inter-8201.pt', map_location=torch.device('cpu')))
```

```python
net
```

```python
trans = transforms.Compose([
    transforms.Resize((256, 256)),
    np.float32,
    transforms.ToTensor(),
    fixed_image_standardization
])
```

```python
data_dir = "C:/Users/U339700/Documents/Palas/dataset/siamesas_resize_train"
paths = []
ims = []
classes = []
for dirpath, dirnames, filenames in os.walk(data_dir):
    for filename in filenames:
        if ".jpg" in filename.lower():
            paths.append(os.path.join(dirpath, filename))
            classes.append(os.path.join(dirpath, filename).split('\\')[1])
            ims.append(filename)
```

```python
batch_size=100
batches = len(paths)//batch_size
len_read = (batches-1)*batch_size+batch_size

embs = []

if len_read < len(paths): # Si el dataset es mayor que el batch size
    for batch in range(batches):        
        print('Batch ', batch+1, ' / ', batches+1)
        b_paths = paths[batch*batch_size:batch*batch_size+batch_size]
        for path in b_paths:
            img1 = Image.open(path)
            img1 = trans(img1)
            img1 = torch.reshape(img1, (1, 3, 256, 256))
            output = net.forward(img1, img1)
            embs.append(output[0][0].tolist())
        # tsne.append(TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs)))
        # embs = []

    # Imgenes ltimo batch
    b_paths = paths[len_read:len(paths)]
    print('Batch ', batches+1, ' / ', batches+1)
    for path in b_paths:
        img1 = Image.open(path)
        img1 = trans(img1)
        img1 = torch.reshape(img1, (1, 3, 256, 256))
        output = net.forward(img1, img1)
        embs.append(output[0][0].tolist())
    # tsne.append(TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs)))
else:
    for batch in range(batches):
        for path in paths:
            img1 = Image.open(path)
            img1 = trans(img1)
            img1 = torch.reshape(img1, (1, 3, 256, 256))
            output = net.forward(img1, img1)
            embs.append(output[0][0].tolist())
        # tsne.append(TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs)))
```

```python
tsne = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(torch.tensor(embs))
tsne_df = pd.DataFrame()
tsne_df['x'] = tsne[:, 0]
tsne_df['y'] = tsne[:, 1]
tsne_df['class'] = classes
tsne_df['img'] = ims
tsne_df
```

```python
sns.set_theme()
sns.lmplot('x', 'y', data=tsne_df, hue='class', fit_reg=False, legend=False, palette="Set2", scatter_kws={"s": 20, 'alpha':0.7})
plt.legend(title='Class', loc='best', labels=['D-0', 'D-2', 'D-3', 'D-4 and D-5'], facecolor='White')
plt.title('Siamese Neural Network image embeddings', fontsize = 14)
plt.show()
```
